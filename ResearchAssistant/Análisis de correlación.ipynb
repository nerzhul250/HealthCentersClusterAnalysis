```python
#pandas for database usage of data
import pandas as pd
#for finding the correlation
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

Se carga los datos sobre el performance de los hospitales

```python
csvPath='rawDataToFormattedData\\PorCentrosDeSalud\\pandaDataPerformance.csv'
hospitales_per=pd.read_csv(csvPath,sep=',')
hospitales_per[hospitales_per.ID==0]
```

```python
csvPath='rawDataToFormattedData\\PorCentrosDeSalud\\pandaData.csv'
hospitales_org=pd.read_csv(csvPath,sep=',')
hospitales_org[hospitales_org.ID==5]
```

Se inicializan las columnas del nuevo dataset

```python
columns=[]
org_num=len(hospitales_org[hospitales_org.ID==0])
per_num=len(hospitales_per[hospitales_per.ID==0])
column_num=per_num+org_num+2
hospitales_num=len(hospitales_per[hospitales_per.EVA=='Acceso_consulta'])
columns_names=[]
columns_names.append('ID')
columns_names.append('HOSPITAL')
for index,row in hospitales_org[hospitales_org.ID==0].iterrows():
    columns_names.append(row['EVA'])
for index,row in hospitales_per[hospitales_per.ID==0].iterrows():
    columns_names.append(row['EVA'].replace(' ','_'))
for col in range(column_num):
    columns.append([])
columns_names
```

```python
for hos in range(hospitales_num):
    hospitales_aux=hospitales_org[hospitales_org.ID==hos]
    columns[0].append(hos);
    columns[1].append(hospitales_aux.iloc[0]['HOSPITAL']);
    i=0
    for index, row in hospitales_aux.iterrows():
        suma=row['TD']+row['MD']+row['ED']+row['NE']+row['DA']+row['MA']+row['TA']
        calificacion=0*row['TD']+row['MD']+2*row['ED']+3*row['NE']+4*row['DA']+5*row['MA']+6*row['TA']
        calificacion=calificacion/suma
        columns[2+i].append(calificacion)
        i+=1
    hospitales_aux=hospitales_per[hospitales_per.ID==hos]
    i=0
    for index, row in hospitales_aux.iterrows():
        suma=row['NU']+row['CN']+row['AV']+row['NE']+row['MV']+row['CS']+row['SI']
        calificacion=0*row['NU']+row['CN']+2*row['AV']+3*row['NE']+4*row['MV']+5*row['CS']+6*row['SI']
        calificacion=calificacion/suma
        columns[org_num+i+2].append(calificacion)
        i+=1
basic_data = dict(zip(columns_names,columns))
hospitales_cal_res = pd.DataFrame (basic_data, columns = columns_names)
hospitales_cal_res
```

```python
corr_matrix=hospitales_cal_res.corr()
corr_matrix=corr_matrix.drop(columns='ID')
corr_matrix=corr_matrix.drop('ID',axis=0)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(corr_matrix.iloc[0:27].loc['Acceso_consulta':])
```

```python
raw_data=hospitales_cal_res.drop(columns='ID').drop(columns='HOSPITAL')
dependent_variable='Coordinación_Atención_en_el_primer_nivel_contra_referencia'
independent_variables=['SAI1 + SAI2 + SAI3 + SAI4 + SAI5',
                       'SAO1 + SAO2 + SAO3 + SAO4',
                       'EI1 + EI2 + EI3',
                       'EC1 + EC2 + EC3 + EC4',
                       'OEO1 + OEO2 + OEO3 + OEO4 + OEO5 + OEO6',
                       'OED1 + OED2 + OED3 + OED4 + OED5']
for i in range(6):
    model = ols(dependent_variable +' ~ '+ independent_variables[i], data=raw_data).fit()
    print(model.rsquared_adj**.5)
    #print (model.rsquared**.5)
    #print (model.params)
    #print (model.summary())
```
